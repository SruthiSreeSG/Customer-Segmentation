**Import the necessary Libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

**Load the Dataset**

```python
dataset = pd.read_excel("/content/Dataset.xlsx")
```

**Print the dataset**

```python
X = dataset.iloc[:, [3,4]].values
print(dataset)
```

**Implement the Elbow Metghod to find the optimal number of Clusters**

```python
wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state= 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss, color='red')
plt.title(' Elbow Method')
plt.xlabel("No of Clusters")
plt.ylabel("WCSS")

plt.show()
```

**Perform K-Means Clustering Algorithm**

```python
kmeans = KMeans(n_clusters = 4, init='k-means++', random_state= 42 )
y_means = kmeans.fit_predict(X)
```

```python
print(y_means)
```

**Scatter plot that depict the no of clusters and their centroid**

```python
plt.scatter(X[y_means == 0, 0], X[y_means == 0, 1], s=100, c='pink', label='C 1')
plt.scatter(X[y_means == 1, 0], X[y_means == 1, 1], s=100, c='grey', label='C 2')
plt.scatter(X[y_means == 2, 0], X[y_means == 2, 1], s=100, c='#00008B', label='C 3')
plt.scatter(X[y_means == 3, 0], X[y_means == 3, 1], s=100, c='#90EE90', label='C 4')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s= 300, c='black', label='Centroids')
plt.title('Cluster Of Amazon Users')
plt.xlabel('Annual Income in INR')
plt.ylabel('Purchase Rating')
plt.legend()
plt.show()
```
